## outlier prediction

```python
df.columns
```

```python
for col in df.columns:
    col_zcore=col+"_zscore"
    df[col_zcore]=(df[col]-df[col].mean())/df[col].std()
```

```python
df['Area Population'].median()
```

```python
df.head()
```

```python
df.columns[6:-1]
```

```python
for col in df.columns[6:-1]:
    col_outlier=col+'_outlier'
    df[col_outlier]=(abs(df[col]>2)).astype(int)
```

## Model

```python
X=df[df.columns[:5]]
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin=LinearRegression()
```

```python
lin.fit(train_x,train_y)
```

```python
test_pred=lin.predict(test_x)
```

```python
train_pred=lin.predict(train_x)
```

```python
r2_score(test_y,test_pred)
```

```python
r2_score(train_y,train_pred)
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cv=cross_val_score(lin,X,y,cv=5)
```

```python
cv.mean()
```
